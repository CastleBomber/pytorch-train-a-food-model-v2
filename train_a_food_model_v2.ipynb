```python
labelA= 'chicken_wings'
labelB= 'chocolate_cake'
#labelA= 'cheesecake'
#labelB= 'ice_cream'
```

```python
# Loop through all the downloaded images
for img in get_image_files(foodPath):
  # Renmae the Images so that the label (chicken_wings or chocolate_cake) is in the file name
  if labelA in str(img):
    img.rename(f"{img.parent}/{labelA}_(img.name)")
  elif labelB in str(img):
    img.rename(f"{img.parent}/{labelB}_(img.name)")
  else: os.remove(img) # If the images are not apart of labelA or labelB

len(get_image_files(foodPath))
```
